## **01.Structured Data**
---------
* Structured data looks like a **table** which is a collection of **rows** and **columns**

```
# please add some big code here

# let's first remove all the features which has massive missing data

print(f'total train data : {housing.shape[0]}')

dict_missing = housing.isna().sum().to_dict()
df_missing = pd.DataFrame.from_dict(data = dict_missing,orient = 'index').reset_index()

df_missing.columns = ['feature','missing_count']

df_missing['per_of_missing_data'] = df_missing['missing_count'] / housing.shape[0]

df_missing.sort_values('per_of_missing_data',ascending = False,inplace = True)

plt.figure(figsize=(20,17))
sns.set_style('darkgrid')
sns.barplot(x = 'per_of_missing_data',y = 'feature', data= df_missing.head(20))
plt.title('Missing percentage of data',fontsize = 18)
plt.show()
```

## **02.Problem of Data Redundancy**
---------
* Redundancy means repetition of data many times in a table
* Redundancy was a big problem in the early stages of database development (during 1970s) when storage space was a big luxury!
* Let's see an example [here](https://docs.google.com/spreadsheets/d/1rTF1uwbFOnptpNFjaRDzaS2ylDTryv2d2jATqVRZ5ys/edit?usp=sharing) to understand the scenarios where redundancy can occur in structured data

## **03.Tackling Redundancy**
---------
* Normalization is the way out
* Other than tackling redundancy, it also helps taking care of the below 3 issues : 
  * **Insertion Anomaly**
  * **Deletion Anomaly**
  * **Updation Anomaly**
