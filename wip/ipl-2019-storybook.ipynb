### **Perception of ordinal data :**
* As we know any data which has a natural order within its elements is referred to as **Ordinal Data**
* With that definition, the IPL table rankings are a natural choice of ordinal variable
* However, when we think of **ranking**, we think of a **staircase** where the lowest ranking entity is at the lowest step and the highest ranking entity stands at the top most one
