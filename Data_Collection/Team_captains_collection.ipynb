```python
base_url = "https://en.wikipedia.org/wiki/Indian_Premier_League#Current_teams"
```

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
source = requests.get(base_url).text
soup = BeautifulSoup(source, 'lxml')

team_table_list = soup.find_all('table')

current_team_list  = team_table_list[2]
former_team_list = team_table_list[3]
print(len(team_table_list))
```

```python
df_current = pd.read_html(str(current_team_list))[0]
df_current = df_current.drop('Team',axis = 1)
df_current.columns = ['Team','City','Home_ground','Debut','Captain','Owner']

df_current['Dissolved'] = 0

df_current
```

```python
df_former = pd.read_html(str(former_team_list))[0]
df_former = df_former.drop('Team',axis = 1)
df_former.columns = ['Team','City','Home_ground','Debut','Dissolved','Owner']

df_former = df_former.drop(0)

df_former['Captain'] = ['Parthiv Patel','Aaron Finch',' Steve Smith','Suresh Raina ']

df_former = df_former[['Team','City','Home_ground','Debut','Captain','Owner','Dissolved']]

df_former
```

```python
df_team_captain_owner = df_current.append(df_former,ignore_index=True)

df_team_captain_owner
```

```python
df_team_captain_owner.to_csv('E:/Google_Drive_Contents/data/team_captain_owner_information.csv',index=False)
```
